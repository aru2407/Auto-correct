```python
import nltk
from nltk.util import ngrams
from nltk.metrics.distance import edit_distance, jaccard_distance
import itertools
from utils import jaccard, levenshtein
```

```python
#data: https://github.com/dwyl/english-words/blob/master/words_alpha.txt
with open("words_alpha.txt","r") as f:
    dictionary = [word.strip() for word in f.readlines()]
```

```python
sentences = ["I am hepy todai",
            "What Ar yo Doin tommorrrow",
            "Can we meeet soonday",
            "It is a beautifool daye"]

print("Text Similarity Using Jaccard Distance")
for sentence in sentences:
    jaccard.jaccard(sentence,dictionary)
    
print("\n\nText Similarity Using Levenshtein/Edit Distance")    
for sentence in sentences:    
    levenshtein.levenshtein(sentence,dictionary)
```
